### Q18. Split and scale the data (2 Mark)

```python
# Split the data into dependent and independent variables
# remove _____ & write the appropriate function variable name
y = pima['Outcome']
x = pima.drop('Outcome', axis = 1)
```

In this code, we used the train_test_split() function from the scikit-learn library to split the data into 
dependent variable y and independent variables X. The target variable 'Outcome' is assigned to y, 
and the rest of the variables are assigned to X by dropping the 'Outcome' column from the pima dataframe 
using the drop() function.

```python
# Using StandarScaler scale the data
# remove _____ & write the appropriate function function name

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(x)
```

In this code, we created an instance of the StandardScaler() function, which is used to standardize the data (i.e., to transform the data so that it has a mean of 0 and standard deviation of 1). We then applied the fit_transform() method to scale the independent variables X. The standardized data is stored in a new variable X_scaled.

```python
# Splitting the data into train and test. Use test size as 15%
# remove _____ & complete the code

from sklearn.model_selection import train_test_split

# Splitting the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=45)
```

In this code, we used the train_test_split() function from the scikit-learn library to split the standardized independent variables (X) and dependent variable (y) into training (x_train and y_train) and testing (x_test and y_test) sets. We set the test_size parameter to 0.15, which means that 15% of the data will be used for testing, and the remaining 85% will be used for training. We also used the stratify parameter to ensure that the proportion of positive and negative outcomes is maintained in both the training and testing sets. Finally, we set the random_state parameter to ensure reproducibility of the results.


### Q19. Train a Logistic Regression Model. What is the Accuracy and how many True Positives and True Negatives did you get? (3 Marks)

```python
# remove _____ & complete the code

# import the model
from sklearn.linear_model import LogisticRegression

# Instantiate the model
lr = LogisticRegression()

# fit the model on the train set
m = lr.fit(x_train, y_train)

# Predict on x_test
y_pred = m.predict(x_test)

# Evaluate the model
print('Accuracy of Logistic Regression model on the train set: {:.2f}'.format(m.score(x_train, y_train)))
print('Accuracy of the Logictic Regression model on the test set: {:.2f}'.format(m.score(x_test, y_test)))

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

### Q20. Train a Random Forest Model. What is the Accuracy and how many True Positives and True Negatives did you get? (3 Marks)

Logistic Regression and Random Forest are two popular classification algorithms used in machine learning. Here are some pros and cons of each model:

Logistic Regression
Pros:
Logistic Regression is a simple and efficient model that can be easily trained and implemented.
It works well when there is a linear decision boundary between the classes, and can handle binary classification problems.
It is interpretable, which means we can understand how the model is making its predictions based on the feature coefficients.
Logistic Regression is less prone to overfitting when compared to other more complex models, such as neural networks.
Cons:
Logistic Regression may not work well when there are non-linear decision boundaries between classes. In such cases, we may need to use more complex models like Random Forest or SVM.
It can only handle binary classification problems out-of-the-box, which means we need to modify the algorithm or use extensions like one-vs-all to handle multi-class classification problems.
Logistic Regression is sensitive to outliers, which can affect the model's performance.
Random Forest
Pros:
Random Forest is a powerful model that can handle both binary and multi-class classification problems effectively.
It is less prone to overfitting due to the way it constructs an ensemble of decision trees.
It can handle non-linear decision boundaries, which makes it suitable for complex classification problems.
Random Forest can handle missing data and noisy data, which makes it more robust than other models.
Cons:
Random Forest can be computationally expensive to train, especially when dealing with large datasets or a large number of trees in the ensemble.
The model can be less interpretable compared to Logistic Regression, making it harder to understand how the model is making its predictions.
It can be sensitive to noisy data, which can affect the model's performance.
In summary, Logistic Regression is a simpler and more interpretable model that can handle binary classification problems efficiently, while Random Forest is a more complex model that can handle both binary and multi-class classification problems and is more robust to missing or noisy data, but can be more computationally expensive to train. The choice of model depends on the specific requirements of the problem at hand, the size and quality of the data, and the trade-offs between accuracy, interpretability, and computational resources.
